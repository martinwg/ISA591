You want find variables that might need **transformations**. Variables that are very skewed can be made more symmetric with a log transformation. E.g.,

`logvar10 = np.log(var10)`  

`logvar10` is more symmetric than the skewed `var10`.

Note: log(0) does NOT exist. Add a negligible amount to `var10` so that the transformation works.

`logvar10 = np.log(var10+0.0001)`

#### **Variable Selection Using Ensembles**

Ensembles calculate the variable importance based on how many times the variables are selected in the splitting of the trees.

```python
## Create the X matrix and the y vector
X = df.drop('SalePrice', axis = 1)
y = df['SalePrice']
```

```python
X.shape
```

```python
## Random Forest (Ensemble)
from sklearn.ensemble import RandomForestRegressor

## instance
rf = RandomForestRegressor()

## fit
rf.fit(X, y)
```

```python
## attribute named variable importance
feat_imp = pd.DataFrame({'Variable': X.columns, 'Importance': rf.feature_importances_})
```

```python
feat_imp.set_index('Variable', inplace = True)
feat_imp.sort_values('Importance', inplace = True, ascending = False)
```

```python
## How many variables to keep?
## 10, 20, 30 (dependent on the problem)
## get a barplot
feat_imp[:20].plot(kind = "barh");
```

```python
## Select top 20 variables
X_final = X[feat_imp[:20].index]
```

```python
X_final.head()
```

```python
## Principal Components allow to create combined variables
## 258 variables (original data set) - 3 PC
from sklearn.decomposition import PCA

## unsupervised technique (we donot need a y variable)
## instance
pca = PCA(n_components = 3) ## select number of components

## fit (unsupervised)
X_PCA = pca.fit_transform(X)
```

```python
X_PCA ## no interpretation, if variables are NOT too correlated does not well
```

```python
## positives: we kept most of the important information of the 258 variables
np.sum(pca.explained_variance_ratio_)
```
