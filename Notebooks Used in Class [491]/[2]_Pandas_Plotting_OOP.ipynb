```python
inv_A = np.random.randint(0, 20, 50)
inv_B = np.random.randint(0, 10, 50)
```

```python
## LINE PLOT
## order, time variable (numeric)
plt.figure(figsize = (10,4))
plt.plot(inv_A, marker = 'o', color = "red", label = "InvA")
plt.plot(inv_B, marker = 'o', color = "blue", label = "InvB")
plt.title('Line Plot of Inventory A')
plt.legend()
plt.show()
```

## **Fit a LR in Python**

```python
X = df[['calories', 'protein']]
y = df.rating
```

```python
## fit (ANY MODEL)

## 1) import the model class
from sklearn.linear_model import LinearRegression

## 2) make an instance (instantiate) the class
myreg = LinearRegression()

## 3) use a method to fit the model ---  .fit()
myreg.fit(X, y)

## 4) predict observation with model --- .predict()
myreg.predict(X)
```
