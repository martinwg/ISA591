## **Unstructured Data: Images**

We are going to look pixel intensities. Intensities of colors are measured in a scale 0-255 (256 intensities). Color images are stored in 3 different channels (Red, Green, Blue). RGB encoding is the most common, but some libraries read images as BRG, BGR.

We need obs and variables.

```python
## cv2 read images
## reads imgs to NumPy images
## BGR (Blue Green Red)
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('/content/Redhawkslogo.png') ## BGR
```

```python
type(img)
```

```python
## obs = 216
## var = 234
## 3 different data sets
img
```

```python
## plot the image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
## Read an img as grayscale (no color 1 channel)
img_gray = cv2.imread('/content/Redhawkslogo.png', cv2.IMREAD_GRAYSCALE)
img_gray
```

```python
## plot rows and cols
plt.imshow(img_gray, cmap = "gray");
```
