```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/martinwg/ISA591/refs/heads/main/data/cereal.csv')
df.head()
```

```python
## X matrix (only predictors)
## SVD, PCA are unsupervised (no response)
X = df.drop(['name', 'mfr', 'type'], axis = 1)
X.head()
```

```python
## SVD (PCA) works the best
## too many predictors
## the predictors are correlated
## relationships with a response not important to describe
```

```python
## SVD
## X = USV'
U, S, V = np.linalg.svd(X, full_matrices=False)
```

```python
## U
## col(0) of U is a linear combination of calories, protein, fat, ....
## col(0) of U contains the MOST information about all 13 variables
## col(1) contains less information than col(0) but NOT included in col(0)
pd.DataFrame(U).head()
```

```python
## V matrix
## recipe to get cols of U (weights of the linear combination)
## components of U
pd.DataFrame(V)
```

```python
## S
## How much information from the 13 columns is contained in each col of U
## S should be a matrix
np.diag(S)
```

### **Decomposition**

```python
pd.DataFrame(U @ np.diag(S) @ V).head()
```

## **THEORY OF PCA**

PCA is a generalization of SVD (it is SVD done on the cov matrix).

```python
## new X with only first 3 cols of X
XR = X.iloc[:, :3]
XR
```

```python
## covariance matrix
## variances on the diagonals and covariances in the off-diagonals
## square and symmetric matrix
XR.cov()
```

```python
## sum the 3 variances
sum_vars = XR["calories"].var() + XR["protein"].var() + XR["fat"].var()
sum_vars
```

```python
print(f'Variance of Calories : {XR["calories"].var()  / sum_vars }')
print(f'Variance of Protein  : {XR["protein"].var() / sum_vars}')
print(f'Variance of Fat      : {XR["fat"].var() / sum_vars}')
```

We want the variables to be **standardized** before we fit PCA, because higher units will mean more information.

```python
XR['calories'].var()
```

```python
## covariance is a measure of how the variables vary together
## -inf to inf (e.g., -1000 will depend on the units but it is negative linear relationship)
from statistics import covariance
cov_calories_protein = covariance(XR['calories'], XR['protein'])
cov_calories_protein
```

```python
## standardized covariance is called correlation
## correlation between calories and protein
## corr = cov /( sd_x1*sd_x2)
## -1 to 1 (standardized)
cov_calories_protein / (XR['calories'].std() * XR['protein'].std())
```

```python
XR.corr()
```

```python
## PCA is a decomposition of the covariance matrix

from sklearn.decomposition import PCA

## instance
pcaanalysis = PCA()
PCs = pcaanalysis.fit_transform(XR)

## how much variation the PCs keep
np.cumsum(pcaanalysis.explained_variance_ratio_)
```

```python
## PCs is actually U
## the cols of U are called principal components
## PC1 contains information about calories, protein, fat
pd.DataFrame(PCs, columns = ['PC1', 'PC2', 'PC3']).head()
```

```python
## The explained variance ratio is just S
pcaanalysis.explained_variance_ratio_
```

```python
## components is the V matrix
pcaanalysis.components_
```

```python
## how is the PC matrix made up?
XR.head(3)
```

```python
pd.DataFrame(PCs, columns = ['PC1', 'PC2', 'PC3']).head(3)
```

```python
## The PC1 (obs 1) is obtained as:
0.99966653*(70 - XR['calories'].mean()) + -0.01088883*(4 - XR['protein'].mean()) +  -0.02341489*(1 - XR['fat'].mean())
```

```python
## The PC1 (obs 2) is obtained as:
0.99966653*(120 - XR['calories'].mean()) + -0.01088883*(3 - XR['protein'].mean()) +  -0.02341489*(5 - XR['fat'].mean())
```

## **Visualization**

When you have too many predictors. You can visualize your data in a scatter plot. That visualization can share insights.

```python
## check malware dataset
df = pd.read_csv('https://raw.githubusercontent.com/martinwg/ISA591/refs/heads/main/data/malware.csv')
df.head()
```

```python
## predicting whether this software is a malware or not
```

```python
## drop missing
df.dropna(inplace = True)
```

```python
## Create an ID
## for visualization we might need an ID, name, identifier
df['ID'] = pd.DataFrame(range(0, df.shape[0]+1   ))
```

```python
## create X matrix
X = df.drop(['Label', 'ID'], axis = 1)
```

```python
X.head()
```

```python
from sklearn.decomposition import PCA

## instance
pcaanalysis = PCA()
PCs = pcaanalysis.fit_transform(X)

## how much variation the PCs keep
np.cumsum(pcaanalysis.explained_variance_ratio_)
```

```python
## SCATTER PLOT OF THE FIRST 2 PCS
## very informative (like a scatter plot of 342 variables)
PCdf = pd.concat([df['ID'], pd.DataFrame(PCs[:, :2], columns = ['PC1', 'PC2'])] , axis = 1)
PCdf.head()
```

```python
PCdf.dropna(inplace = True)
```

```python
import plotly.express as px

#cereals.reset_index(inplace = True) # to avoid different indices
#PCdf = pd.concat([cereals['name'], pd.DataFrame(PCs[:,0:2], columns = ['PC1', 'PC2'])], axis = 1)

## change name for ID
fig = px.scatter(PCdf, x="PC1", y="PC2", text="ID", log_x=False, size_max=100, color= "ID")
fig.update_traces(textposition='top center')
fig.update_layout(title_text='Cereal Brands and Principal Components', title_x=0.5)
fig.show()
```
