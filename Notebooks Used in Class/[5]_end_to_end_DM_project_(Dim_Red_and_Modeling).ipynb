You want to find variables that might need transformations. Variables that are very skewed might need a log transformation, i.e,

`logvar10 = np.log(var10)`

Note: log(0) does not exist. So adding a negligible amount to `var10` removes the problem, i.e.,

`logvar10 = np.log(var10+0.001)`

```python
df.PoolQC.fillna('No Pool', inplace = True)
```

```python
## for loop to assign value "Unknown" to ALL categorical vars with missing vals
for i in df[(df.isna().sum() > 0).index].select_dtypes('O'):
  df[i].fillna('Unknown', inplace = True)
```

```python
## for loop to assign mean or median to ALL numeric vars with missing vals
for i in df[(df.isna().sum() > 0).index].select_dtypes(['int', 'float']):
  df[i].fillna(df[i].median(), inplace = True)
```

```python
# prompt: Which variables are the most correlated?

# Calculate the correlation matrix
corr_matrix = df.corr()

# Find the most correlated variables
most_correlated = corr_matrix.abs().unstack().sort_values(ascending=False)

# Filter out the diagonal (correlation of a variable with itself)
most_correlated = most_correlated[most_correlated != 1]

# Print the top 10 most correlated pairs
print(most_correlated[:50])
```

#### **Variable selection using Ensemble**

Fit a random forest and get the importance of each variable. Rank those.

```python
## Create the X matrix and y vector
X = df.drop('SalePrice', axis = 1)  ## X Matrix
y = df.SalePrice                    ## y vector
```

```python
## fit a RF (default)
from sklearn.ensemble import RandomForestRegressor

## instance
rf = RandomForestRegressor()

## fit
rf.fit(X, y)
```

```python
## how important are variables
var_importance = rf.feature_importances_
```

```python
## put var_importance in a data frame and sort it
var_imp_df = pd.DataFrame({'Variable': X.columns, 'Importance': var_importance})
var_imp_df.sort_values(by = 'Importance', ascending = False, inplace = True)
```

```python
var_imp_df[:11]
```

```python
## getting a bar chart of importance
var_imp_df.iloc[:30].plot(kind = "barh")
plt.title("Feature Importance")
plt.show()
```

```python
## Create a reduced data set with ONLY the most important variables
## how many vars? dependent on the problem
## Let's select only the top 20
X_final = X[var_imp_df.iloc[:20].Variable]
```

```python
X_final.head()
```

```python
## if variables are too correlated
## PCA would be helpful
from sklearn.decomposition import PCA

## PCA is an unsupervised learning algorithm
## NOT GOING TO ASK FOR Y

## instance
pca = PCA(n_components=3)
X_PCA = pca.fit_transform(X)
```

```python
## what percentage of the information included in 259 variables do WE keep
## this works well if there are really big correlations among most variables
np.sum(pca.explained_variance_ratio_)
```

```python
X_PCA
```

```python
## 258 predictors
## 1st PC keeps 98.28% of the variability of the 258 predictors
pca.explained_variance_ratio_
```

```python
PC_df = pd.DataFrame(X_PCA,  columns = ['PC1', 'PC2', 'PC3'])
PC_df
```

```python
# prompt: Create a 3D scatter plot of PC_df

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(PC_df['PC1'], PC_df['PC2'], PC_df['PC3'])

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')

plt.title('3D Scatter Plot of Principal Components')
plt.show()
```

#### 1) Linear Regression

Family of models. Options: quadratic terms, interactions, drop insignificant based on p-values.

```python
## Predictive model
## not checking on p-values
## not checking on goodness of fit metrics (R2, RMSE, F-statistic, assumptions)

## import
from sklearn.linear_model import LinearRegression

## make an instance (instantiate)
lr = LinearRegression()

## fit on the training data
lr.fit(X_train, y_train)
```

```python
## Neural Network
## Deep NN (3 hidden layers)
## 50 neuron, 25 neurons, 10

## import
from sklearn.neural_network import MLPRegressor

## instance
nn = MLPRegressor(hidden_layer_sizes = (50, 25, 10), max_iter = 2000)

## fit
nn.fit(X_train, y_train)
```
